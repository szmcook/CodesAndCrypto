```python
import time

def timingStatistics(n):
    N = 10 ** 6

    R = 0
    
    duration = []
    for y in range(n):
        durationy = []
        for s in range(n):
            start = time.time()
            for i in range(N):
                R = (s * y) % n
            end = time.time()
            computationTime = int( 10 ** 6 * (end - start) )
            
            durationy.append(computationTime)
        duration.append(durationy)
        
    return duration

def printTimingStatistics(n):
    duration = timingStatistics(n)

    for i in range(n):
        print(duration[i], '\n')

    return duration
```
